```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd

import gseapy as gp
from gseapy.plot import barplot, dotplot
```

```python
name = 'Reprogramming'
rep = pd.read_csv('reprogramming_full.csv', index_col=0)
rnk = rep[['symbol', 'logFC']].reset_index().drop('index', 1)
rnk.columns = [0, 1]
```

```python
gene_sets=['GO_Biological_Process_2021', 'KEGG_2019_Mouse', 'Reactome_2016']
```

```python
gsea = gp.prerank(rnk=rnk,
                gene_sets='KEGG_2019_Mouse',
                processes=4,
                outdir='tmp/enrichment/%s' % name,
                permutation_num=10000, # reduce number to speed up testing
                format='png', seed=3
                )

res = gsea.results    
```
