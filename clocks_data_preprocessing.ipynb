# Mouse

```r
df <- read.csv('clocks/mouse_full_data_29ds.csv', check.names=FALSE, row.names=1)
```

```r
time <- df$TIME
df <-df[,!names(df) %in% c("TIME")]
```

```r
source('utils/FUN.Align_to_normal_dist.R')
source('utils/FUN.Preprocessing_for_clock.R')
avg <- read.csv('clocks/Average_gene_expression.csv', row.names=1)
```

```r
fin <- preprocessing_for_clock(t(df), avg)
```

```r
ids <- rownames(fin)
pre <- data.frame()
for (i in 1:length(ids)){
    pre_id <- ids[[i]]
    id <- pre_id
    id <- sub('\\$\\$', '\\$', id)
    id <- sub('\\$', '\\$', id)
    s <- strsplit(id, "\\$")[[1]]
    gse <- s[[1]]
    treat <- paste(s[2:(length(s)-1)], collapse="_")
    gsm <- s[[length(s)]]

    pre <- rbind(pre, data.frame(list("ID"=pre_id, "GSE"=gse, "Treatment"=treat, "GSM"=gsm)))
    cat(gse, treat, gsm, '\n')
}
pre$Time <- time

pre$Treatment <- gsub("OKSM", "OSKM", pre$Treatment)
pre$Treatment <- gsub("OKMS", "OSKM", pre$Treatment)
```

```r
tr <- pre$Treatment

BatchTreatment <- c()
for (t in tr){
    if (grepl("OSKM" ,t)){
        BatchTreatment <- c(BatchTreatment, "OSKM")
    }else if(grepl("OSK" ,t)){
        BatchTreatment <- c(BatchTreatment, "OSK")
    }else if(grepl("GFP" ,t)){
        BatchTreatment <- c(BatchTreatment, "Control")
    }else{
        BatchTreatment <- c(BatchTreatment, "7F")
    }
}

pre$BatchTreatment <- BatchTreatment
pre$BatchTreatment[pre$Treatment == 'OK+9MS'] <- "OSKM"
pre$BatchTreatment[pre$Treatment == 'Oct4+SK'] <- "OSK"
pre$BatchTreatment[pre$Treatment == 'Oct6+SK'] <- "OSK"
pre$BatchTreatment[pre$Treatment == 'Oct4+defSox+K'] <- "OSK"
pre$BatchTreatment[pre$Treatment == 'OSKM-dox_mef'] <- "Control"
```

# Human

```r
#human
annot <- read.csv('clocks/human_data_annotation.csv', check.names=F)
annot$ID <- gsub(".GSM\\d+", "", annot$ID)
```

```r
tmp <- as.character(annot$Treatment)

annot$Tissue <- gsub("OSKM.", "", tmp)

annot$Tissue <- gsub('AdMSC', "ASC", annot$Tissue)
annot$Tissue <- gsub('NHBE', "BEC", annot$Tissue)
annot$Tissue <- gsub('HASC', "ASC", annot$Tissue)
annot$Tissue <- gsub('HPrEC', "PrEC", annot$Tissue)
annot$Tissue <- gsub('HBEC', "BEC", annot$Tissue)
annot$Tissue <- gsub('HDF', "DF", annot$Tissue)
annot$Tissue <- gsub('HFF', "FF", annot$Tissue)
annot$Tissue <- gsub('HA', "A", annot$Tissue)
annot$Treatment <- "OSKM"
```
